```python
def get_trials(
    logger, 
    cursorOn=True, 
    feedbackLinked=True, 
    rewardOn=True, 
    success=True,
):

    trials_run = ~np.isnan(logger['trials']['frameNum_trialEnd'].to_numpy())
    
    trials_typed = np.array([True] * logger['trials'].shape[0])
    trials_typed = trials_typed * (logger['trials']['trialType_cursorOn'].to_numpy().astype(bool) == cursorOn) if cursorOn is not None else trials_typed
    trials_typed = trials_typed * (logger['trials']['trialType_feedbackLinked'].to_numpy().astype(bool) == feedbackLinked) if feedbackLinked is not None else trials_typed
    trials_typed = trials_typed * (logger['trials']['trialType_rewardOn'].to_numpy().astype(bool) == rewardOn) if rewardOn is not None else trials_typed
    trials_typed = trials_typed * (logger['trials']['success_outcome'].to_numpy().astype(bool) == success) if success is not None else trials_typed

    trials_typed = np.nan_to_num(trials_typed, nan=0)[trials_run].astype(bool)

    frames_trialEnds   = (logger['trials']['frameNum_trialEnd'].to_numpy() - 1)[trials_run].astype(np.int64)
    frames_trialStarts = (logger['trials']['frameNum_trialStart'].to_numpy() - 1)[trials_run].astype(np.int64)

    framesEnd_typed = frames_trialEnds[trials_typed]
    framesStart_typed = frames_trialStarts[trials_typed]
    
    out = {
        'trials_typed': trials_typed,
        'trialNum': logger['trials']['trialNum_trialStart'][trials_run].to_numpy().astype(np.int64),
        'frames_start': frames_trialStarts,
        'frames_end': frames_trialEnds,
        'frames_start_typed': framesStart_typed,
        'frames_end_typed': framesEnd_typed,
        'duration': (logger['trials']['frameNum_trialEnd'] - logger['trials']['frameNum_trialStart']).to_numpy()[trials_run],
        'duration_typed': framesEnd_typed - framesStart_typed
    }
    
    return out
```

```python
def prep_array_for_regression(arr, mean_sub=False, std_div=False, device='cpu'):
    out = torch.as_tensor(arr, dtype=torch.float32, device=device)
    if mean_sub:
        out = out - torch.nanmean(out, dim=1, keepdim=True)
    out = torch.nan_to_num(out, nan=0.0)
    if std_div:
        out = out / out.std(1, keepdim=True)
    return out
```

```python
def dict_to_numpy(d):
    out = {}
    for key, val in d.items():
        if isinstance(val, dict):
            out[key] = dict_to_numpy(val)
        elif isinstance(val, torch.Tensor):
            out[key] = val.cpu().numpy()
        elif isinstance(val, np.ndarray):
            out[key] = val
        else:
            out[key] = val

    return out
```

```python
crossmodel_outs = np.zeros((len(regression_outs), len(regression_outs)), dtype=object)
for ii, date_i in tqdm(enumerate(list(regression_outs.keys()))):
    X = prep_array_for_regression(dFoFs[date_i].T)
    y_true = prep_array_for_regression(tcas[date_i])
    
    for jj, date_j in enumerate(regression_outs.keys()):
        model = lambda X, theta, bias: X @ theta + bias
    
        theta, bias = prep_array_for_regression(regression_outs[date_j]['thetas']), prep_array_for_regression(regression_outs[date_j]['biases'])
        y_rec = model(X, theta, bias)
#         y_rec = regression_outs[date_j]['y_recs']
    
        v1_orth, EVR, EVR_total_weighted, EVR_total_unweighted = bnpm.similarity.pairwise_orthogonalization_torch(
            v1=y_true,
            v2=y_rec,
            center=True,
            device=DEVICE_regression,
        )
    
        crossmodel_outs[ii, jj] = {}
        crossmodel_outs[ii, jj]['thetas'] = thetas
        crossmodel_outs[ii, jj]['y_recs'] = y_recs
        crossmodel_outs[ii, jj]['biases'] = biases
        crossmodel_outs[ii, jj]['orthogonalization'] = {
            'v1_orth': v1_orth, 
            'EVR': EVR, 
            'EVR_total_weighted': EVR_total_weighted, 
            'EVR_total_unweighted': EVR_total_unweighted, 
        }
        crossmodel_outs[ii, jj] = dict_to_numpy(crossmodel_outs[ii, jj])
```

```python
rank = crossmodel_outs[0,0]['thetas'].shape[1]
```

```python
EVRs = np.zeros(tuple(crossmodel_outs.shape) + (rank,), dtype=np.float32)
for ii in range(EVRs.shape[0]):
    for jj in range(EVRs.shape[1]):
        EVRs[ii,jj] = crossmodel_outs[ii,jj]['orthogonalization']['EVR']
        # test2[ii,jj] = crossmodel_outs[ii,jj]['orthogonalization']['EVR'][3]
```
