```python
def register_FOVs(templateFOV, FOVs, ROIs, return_sparse=False):
    dims = templateFOV.shape
    x_grid, y_grid = np.meshgrid(np.arange(0., dims[1]).astype(np.float32), np.arange(0., dims[0]).astype(np.float32))

    template_norm = np.uint8(templateFOV * (templateFOV > 0) * 255)
    
    for ii in range(len(FOVs)):
        FOV_norm = np.uint8(FOVs[ii] * (FOVs[ii] > 0) * 255)
        flow = cv2.calcOpticalFlowFarneback(np.uint8(template_norm * 255), np.uint8(FOV_norm * 255), None,
                                            0.5, 3, 128, 3, 7, 1.5, 0)
        x_remap = (flow[:, :, 0] + x_grid).astype(np.float32)
        y_remap = (flow[:, :, 1] + y_grid).astype(np.float32)

        # A_2t = np.reshape(A2, dims + (-1,), order='F').transpose(2, 0, 1)
#         A_2t = sf_all[6]
        ROI_aligned = np.stack([cv2.remap(img.astype(np.float32), x_remap, y_remap, cv2.INTER_NEAREST) for img in ROIs[ii]], axis=0)
        # A2 = np.reshape(A2.transpose(1, 2, 0), (A1.shape[0], A_2t.shape[0]), order='F')
        
        if return_sparse:
            yield scipy.sparse.csc_matrix(ROI_aligned.reshape(ROI_aligned.shape[0], -1))
        else:
            yield ROI_aligned
```

```python
def make_block_batches(
    frame_height=512,
    frame_width=1024,
    block_height=100,
    block_width=100,
    overlap_tileMultiplier=2,
    inner_block_height=None,
    inner_block_width=None,
):     
    
    # block prep
    block_height_half = block_height//2
    block_width_half = block_width//2
    
    # inner block prep
    if inner_block_height is None:
        inner_block_height = block_height * 0.5
    if inner_block_width is None:
        inner_block_width = block_width * 0.5
        
    inner_block_height_half = inner_block_height//2
    inner_block_width_half = inner_block_width//2
    
    # find centers of blocks
    n_blocks_x = frame_width / block_width
    n_blocks_x = np.ceil(n_blocks_x*overlap_tileMultiplier).astype(np.int64)
    centers_x = np.linspace(
        start=block_width_half,
        stop=frame_width - block_width_half,
        num=n_blocks_x,
        endpoint=True
    )

    n_blocks_y = frame_height / block_height
    n_blocks_y = np.ceil(n_blocks_y*overlap_tileMultiplier).astype(np.int64)
    centers_y = np.linspace(
        start=block_height_half,
        stop=frame_height - block_height_half,
        num=n_blocks_y,
        endpoint=True
    )
    
    # make blocks
    blocks, inner_blocks = [], []
    for i_x in range(n_blocks_x):
        for i_y in range(n_blocks_y):
            blocks.append([
                list(np.int64([centers_y[i_y] - block_height_half , centers_y[i_y] + block_height_half])),
                list(np.int64([centers_x[i_x] - block_width_half , centers_x[i_x] + block_width_half]))
            ])
            
            inner_blocks.append([
                list(np.int64([centers_y[i_y] - inner_block_height_half , centers_y[i_y] + inner_block_height_half])),
                list(np.int64([centers_x[i_x] - inner_block_width_half , centers_x[i_x] + inner_block_width_half]))                
            ])
    return blocks, inner_blocks, (centers_y, centers_x)
```

```python
## Process inner block (ROI idx ownership)
idx_roi_inInnerBlock = []
for ii, block in enumerate(tqdm(inner_blocks)):
    sf_block_cat = np.concatenate([sfs.toarray().reshape(sfs.shape[0], frame_height, frame_width)[:,block[0][0]:block[0][1], block[1][0]:block[1][1]] for sfs in ROIs_aligned], axis=0) # cropped spatial footprints. shape=(n_rois across all sessions, inner_block_height, inner_block_width)
    idx_outOfInnerBlock = np.sum(sf_block_cat, axis=(1,2)) == 0 # roi idx bool that have no pixels in the current block
    idx_roi_inInnerBlock.append(idx_outOfInnerBlock==False) # roi idx bool that have any pixels in the current block
    
```

Heuristic to determine which block to trust for each ROI:

- for each block, determine how many rois are a part of each cluster
- for each roi, determine how many 'neighbor' rois are in it's cluster


- then, for each roi, trust the block that clustered it with the greatest number of ROIs that wasn't too many (ranges between a user defined range)

```python
roi_labelFreqs = np.ones((len(block_rois), len(block_rois[0]['idx_roi_inFrame']))) * np.nan
for i_block, block in enumerate(tqdm(block_rois)):
    labels = block['db'].labels_
    labels_unique = np.unique(labels)
    label_sizes = [list(labels).count(label) for label in labels_unique]
    for i_roi in range(len(block_rois[0]['idx_roi_inFrame'])):
        idx_outer = np.where(block_rois[i_block]['idx_roi_inFrame'])[0]
        if i_roi in idx_outer: # check if roi was in the outer frame
            arg_outer = np.where(i_roi == idx_outer) # get argument in outer list of this roi
            if i_roi in np.where(idx_roi_inInnerBlock[i_block])[0]: # check if roi was in the inner frame
                label = labels[arg_outer] # get the label of this roi in this block
                arg_unique = np.where(labels_unique == label)[0][0]
                size = label_sizes[arg_unique]
                
                roi_labelFreqs[i_block, i_roi] = size # shape = (n_blocks, n_roi total), values= n_roi in cluster for a given roi in a given block
```

```python
%matplotlib notebook
plt.figure()
# plt.imshow(labels_allInnerBlocks.T, aspect='auto', interpolation='none')
plt.imshow(roi_labelFreqs.T, aspect='auto', interpolation='none', vmax=7)
```

```python
roi_labelFreqs_noNaN = copy.copy(roi_labelFreqs)
roi_labelFreqs_noNaN[np.isnan(roi_labelFreqs_noNaN)] = 0
roi_labelFreqs_noNaN = roi_labelFreqs_noNaN.astype(np.int64)
```

```python
clusterSizeRange = [7,7]
roi_labelFreqs_ranged = roi_labelFreqs_noNaN * (roi_labelFreqs_noNaN<=clusterSizeRange[1]) * (roi_labelFreqs_noNaN>=clusterSizeRange[0])
roiBestBlock = np.argmax(roi_labelFreqs_ranged, axis=0)
roiBestBlock[roi_labelFreqs_ranged.max(0) == 0] = -1
```

now go into the 'best block' for each roi and get it's label

```python
labels_best = np.ones(len(block_rois[0]['idx_roi_inFrame'])) * -2 + 1j*0
for i_roi, bb in enumerate(tqdm(roiBestBlock)):
    if bb != -1:
        idxAll_in_bb = np.where(block_rois[bb]['idx_roi_inFrame'])[0]
        idx_in_bb = np.where(i_roi == idxAll_in_bb)[0]
        labels_best[i_roi] = block_rois[bb]['db'].labels_[idx_in_bb] + 1j*bb
```

```python
labels_best_uniques = np.unique(labels_best)
labels_cat_final = np.ones(len(labels_best), dtype=np.int64) * -1
for ii in range(len(labels_best_uniques)):
    labels_cat_final[np.where(labels_best == labels_best_uniques[ii])[0]] = ii
```
