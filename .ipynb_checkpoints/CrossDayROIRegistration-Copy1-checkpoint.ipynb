```python
def register_FOVs(templateFOV, FOVs, ROIs, return_sparse=False):
    dims = templateFOV.shape
    x_grid, y_grid = np.meshgrid(np.arange(0., dims[1]).astype(np.float32), np.arange(0., dims[0]).astype(np.float32))

    template_norm = np.uint8(templateFOV * (templateFOV > 0) * 255)
    
    for ii in range(len(FOVs)):
        FOV_norm = np.uint8(FOVs[ii] * (FOVs[ii] > 0) * 255)
        flow = cv2.calcOpticalFlowFarneback(np.uint8(template_norm * 255), np.uint8(FOV_norm * 255), None,
                                            0.5, 3, 128, 3, 7, 1.5, 0)
        x_remap = (flow[:, :, 0] + x_grid).astype(np.float32)
        y_remap = (flow[:, :, 1] + y_grid).astype(np.float32)

        # A_2t = np.reshape(A2, dims + (-1,), order='F').transpose(2, 0, 1)
#         A_2t = sf_all[6]
        ROI_aligned = np.stack([cv2.remap(img.astype(np.float32), x_remap, y_remap, cv2.INTER_NEAREST) for img in ROIs[ii]], axis=0)
        # A2 = np.reshape(A2.transpose(1, 2, 0), (A1.shape[0], A_2t.shape[0]), order='F')
        
        if return_sparse:
            yield scipy.sparse.csc_matrix(ROI_aligned.reshape(ROI_aligned.shape[0], -1))
        else:
            yield ROI_aligned
```

```python
def make_block_batches(
    frame_height=512,
    frame_width=1024,
    block_height=100,
    block_width=100,
    overlap_tileMultiplier=2,
    inner_block_height=None,
    inner_block_width=None,
):     
    
    # block prep
    block_height_half = block_height//2
    block_width_half = block_width//2
    
    # inner block prep
    if inner_block_height is None:
        inner_block_height = block_height * 0.5
    if inner_block_width is None:
        inner_block_width = block_width * 0.5
        
    inner_block_height_half = inner_block_height//2
    inner_block_width_half = inner_block_width//2
    
    # find centers of blocks
    n_blocks_x = frame_width / block_width
    n_blocks_x = np.ceil(n_blocks_x*overlap_tileMultiplier).astype(np.int64)
    centers_x = np.linspace(
        start=block_width_half,
        stop=frame_width - block_width_half,
        num=n_blocks_x,
        endpoint=True
    )

    n_blocks_y = frame_height / block_height
    n_blocks_y = np.ceil(n_blocks_y*overlap_tileMultiplier).astype(np.int64)
    centers_y = np.linspace(
        start=block_height_half,
        stop=frame_height - block_height_half,
        num=n_blocks_y,
        endpoint=True
    )
    
    # make blocks
    blocks, inner_blocks = [], []
    for i_x in range(n_blocks_x):
        for i_y in range(n_blocks_y):
            blocks.append([
                list(np.int64([centers_y[i_y] - block_height_half , centers_y[i_y] + block_height_half])),
                list(np.int64([centers_x[i_x] - block_width_half , centers_x[i_x] + block_width_half]))
            ])
            
            inner_blocks.append([
                list(np.int64([centers_y[i_y] - inner_block_height_half , centers_y[i_y] + inner_block_height_half])),
                list(np.int64([centers_x[i_x] - inner_block_width_half , centers_x[i_x] + inner_block_width_half]))                
            ])
    return blocks, inner_blocks, (centers_y, centers_x)
```

this is going to be a pain in the ass, but should work:
- each block will have an inner block
- each block will be assigned all the rois that have any pixels within the inner block
    - this will result in rois being assigned to multiple different blocks
- after clustering, each block will compared to every other block to find redundant rois
    - each block will compete with eachother pairwise for rois. Whichever block has a bigger cluster containing that roi gets it
        - each ROI is assigned to the block where it is found to be a part of the largest cluster that is less than n_sessions
        
        
- just loop through each roi and then loop through each block and recover what cluster it is a part of and then how big that cluster is. store that value pair (block, cluster size)
- then each ROI is assigned the block with the largest cluster under n_sessios
- then populate each block only with ROIs that belong to that block
